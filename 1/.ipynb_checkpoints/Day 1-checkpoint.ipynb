```python
import math
```

```python
with open('input.txt') as f:
    input = f.read()
```

```python
input
```

```python
parts_s = input.split()
```

```python
parts_s
```

```python
parts_int = [int(part) for part in parts_s]
print(parts_int)
```

```python
def fuel_for_part(part):
    return math.floor(part / 3) - 2
```

```python
fuels = [fuel_for_part(part) for part in parts_int]
print(fuels)
```

```python
total_fuel = sum(fuels)
print(total_fuel)
```
